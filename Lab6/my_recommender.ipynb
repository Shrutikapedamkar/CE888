```python
#mean of each movie
joke_mean =df.mean(axis=0)
joke_mean
```

```python
#movie with highest ratings
high_rated_joke= joke_mean[joke_mean == max(joke_mean)]
high_rated_joke
```

```python
#movie with worst rating
worst_rated_joke= joke_mean[joke_mean == min(joke_mean)]
worst_rated_joke
```

```python
print("Best joke: ",high_rated_joke.index.get_level_values(0)[0] )
print("Worst joke?: ",worst_rated_joke.index.get_level_values(0)[0] )
```
